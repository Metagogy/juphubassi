**Question**: Load, Split, normalize and flatten the mnist dataset. Build, compile, train and evaluate the model using SGD optimizer, loss as "mean_squared_error".


**Description :** 

Load the mnist dataset using inbuilt dataset function and split into training and testing images.

Normalize the dataset images by dividing it with 255 and subtract 0.5 of train and test images. 

Flatten the dataset images by using reshape. 

Build the mnist dataset model by using a sequential layer and adding the first dense layer with 64 neurons and relu activation. Second layer has 64 neurons with softmax activation and output layer as sigmoid activation with 10 neurons.

Compile the model using ‘sgd’ optimizer, loss as "mean_squared_error" and train the model using 10 epochs, 32 as batch size.

Evaluate the mnist model by using test images and labels.

**Level:** Hard

**Input format :** 

Mnist dataset

**Output format :**

Neural network model accuracy

**Sample input :**

Load the dataset using inbuilt function

**Sample Output :**

Accuracy : 80


```python
import tensorflow as tf
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

# Normalize the images.
train_images = (train_images / 255) - 0.5
test_images = (test_images / 255) - 0.5

# Flatten the images.
train_images = train_images.reshape((-1, 784))
test_images = test_images.reshape((-1, 784))


# And to the following:
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(64, activation = 'relu' , input_shape=(784, )))
model.add(tf.keras.layers.Dense(64, activation = 'softmax' ))
model.add(tf.keras.layers.Dense(10, activation = 'sigmoid' ))
model.summary()

# Compile the model.
model.compile('SGD', loss=tf.keras.losses.MeanSquaredError() ,  metrics = ['accuracy'] )
```

```python
# Train the model.
model.fit(
  train_images,
  tf.keras.utils.to_categorical(train_labels),
  epochs=10,
  batch_size=32,
)
```

```python
# Evaluate the model.
model.evaluate(
  test_images,
  tf.keras.utils.to_categorical(test_labels)
)
```
